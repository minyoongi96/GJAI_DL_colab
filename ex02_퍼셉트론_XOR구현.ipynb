## XOR 퍼셉트론 구현
- XOR 연산결과를 예측할 수 있는 퍼셉트론을 구현해보기
- XOR은 입력값을 OR연산한 결과와 NAND연산한 결과를 AND연산하면 구할 수 있다.
- XOR = (X1 NAND X2) AND (X1 OR X2)

<br>

### 활성화 함수 만들기
- 각 퍼셉트론에 적용할 활성화함수를 step function(계단 함수)로 구현해보기
- 임계치를 0으로 설정
- 임계치보다 높으면 1, 낮으면 0으로 반환하기

```python
import numpy as np
```

```python
# 계단 함수, 임계치=0
def activation_func(x):
    if x <= 0:
        return 0
    else:
        return 1
```

### AND 퍼셉트론 만들기
- 퍼셉트론을 선형 모델 + 활성화함수로 구현
- 활성화 함수를 통과하면 AND연산 결과가 나오는 가중치와 절편으로 선형 모델 생성

```python
def AND(x1, x2):
    w1, w2, b = 1, 1, -1.5
    y = w1*x1 + w2*x2 + b

    return activation_func(y)
```

```python
print(AND(1, 1))
print(AND(1, 0))
print(AND(0, 1))
print(AND(0, 0))
```

### OR 퍼셉트론 만들기

```python
def OR(x1, x2):
    w1, w2, b = 1, 1, 0
    y = w1*x1 + w2*x2 + b
    
    return activation_func(y)
```

```python
print(OR(1, 1))
print(OR(1, 0))
print(OR(0, 1))
print(OR(0, 0))
```

#### NAND 퍼셉트론 만들기

```python
def NAND(x1, x2):
    w1, w2, b = -1, -1, 1.5
    y = w1*x1 + w2*x2 + b
    
    return activation_func(y)
```

```python
print(NAND(1, 1))
print(NAND(1, 0))
print(NAND(0, 1))
print(NAND(0, 0))
```

```python
def XOR(x1, x2):
    a1 = NAND(x1, x2)
    a2 = OR(x1, x2)

    return AND(a1, a2)
```

```python
print(XOR(1, 1))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(0, 0))
```
